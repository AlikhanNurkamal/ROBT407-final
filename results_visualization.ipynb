```python
import matplotlib.pyplot as plt


models = ['CCT-3-2', 'CCT-7-2', 'CCT-14-2', 
          'CCT-3-3', 'CCT-7-3', 'CCT-14-3', 
          'ViT-Lite-16', 'ViT-Lite-32', 'ResNet50', 'ResNet101']

params = [1.505, 4.505, 21.927, 2.509, 5.710, 23.533, 3.302, 3.716, 25.6, 44.5]

scores = [1.4769, 1.9212, 1.6867, 1.2841, 1.5540, 1.4948, 3.0401, 3.5295, 1.6890, 1.1834]

colors = ['blue', 'blue', 'blue', 'red', 'red', 'red', 'yellow', 'yellow', 'green', 'green']

plt.figure(figsize=(10, 6))

for i in range(2):
    plt.scatter(params[i], scores[i], color=colors[i])
    plt.plot(params[i:i+2], scores[i:i+2], color=colors[i])

plt.scatter(params[2], scores[2], color=colors[2])
    
for i in range(3, 5):
    plt.scatter(params[i], scores[i], color=colors[i])
    plt.plot(params[i:i+2], scores[i:i+2], color=colors[i])

plt.scatter(params[5], scores[5], color=colors[5])

for i in range(6, 7):
    plt.scatter(params[i], scores[i], color=colors[i])
    plt.plot(params[i:i+2], scores[i:i+2], color=colors[i])

plt.scatter(params[7], scores[7], color=colors[7])

for i in range(8, 9):
    plt.scatter(params[i], scores[i], color=colors[i])
    plt.plot(params[i:i+2], scores[i:i+2], color=colors[i])

plt.scatter(params[9], scores[9], color=colors[9])

plt.title('Kaggle Score vs Number of Parameters')
plt.xlabel('Number of Parameters (M)')
plt.ylabel('Kaggle Score')
plt.grid(True)

for i, model in enumerate(models):
    plt.annotate(model, (params[i], scores[i]), fontsize=8, alpha=1)

plt.show()
```
