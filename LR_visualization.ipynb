```python
import matplotlib.pyplot as plt
import numpy as np
import math

config = {'LR_INIT': 0.003, 'EPOCHS': 100}
warmup_ep = 10
warmup = True  
enable_cos = True  

epochs = np.arange(config['EPOCHS'])

lr_values = np.zeros(config['EPOCHS'])
for epoch in epochs:
    lr = config['LR_INIT']
    if warmup and epoch < warmup_ep:
        lr = lr / (warmup_ep - epoch)
    elif enable_cos:
        lr *= 0.5 * (1. + math.cos(math.pi * (epoch - warmup_ep) / (config['EPOCHS'] - warmup_ep)))
    lr_values[epoch] = lr

plt.plot(epochs, lr_values)
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedule')
plt.grid(True)
plt.show()
```
