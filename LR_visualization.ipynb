```python
from config import config
import matplotlib.pyplot as plt
import numpy as np
import math

warmup = True
enable_cos = True

epochs = np.arange(config['EPOCHS'])

lr_values = np.zeros(config['EPOCHS'])
for epoch in epochs:
    lr = config['LR_INIT']
    if warmup and epoch < config['WARMUP_EPOCHS']:
        lr = lr / (config['WARMUP_EPOCHS'] - epoch)
    elif enable_cos:
        lr *= 0.5 * (1. + math.cos(math.pi * (epoch - config['WARMUP_EPOCHS']) / (config['EPOCHS'] - config['WARMUP_EPOCHS'])))
    lr_values[epoch] = lr

plt.plot(epochs, lr_values)
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedule')
plt.grid(True)
plt.show()
```
